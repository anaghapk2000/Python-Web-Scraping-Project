### ***Web scraping Laptop details from flipCart website for creating a CSV file and store the data into a SQLite database***

```python
# importing required libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
import sqlite3 as sq
```

```python
# Enter the number of pages you want to scrap from flipCart

Pages=int(input("Enter number of pages you want to scrap: "))
```

```python
# Initializing lists for storing scraped datas

product_name=[]
product_price=[]
product_rating=[]
product_details=[]


# webscrapping the details about laptops from flipCart

for page in range(1,Pages+1):
  
  URL = 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page='
  
  req = requests.get(URL + str(page))
  content=BeautifulSoup(req.content,'html.parser')
  #print(content)
  name=content.find_all('div',{"class":"_4rR01T"})
  price=content.find_all('div',{"class":"_30jeq3 _1_WHN1"})   
  rating=content.find_all('div',{"class":"_3LWZlK"})
  details=content.find_all('ul',{"class":"_1xgFaf"})

  for page in name:
    product_name.append(page.text)   

  for page in price:
    product_price.append(page.text)

  for page in rating:
    product_rating.append(page.text)

  for page in details:
    product_details.append(page.text)
```

```python
print("Number of Laptops = ",len(product_name))
```

```python
# list view of Laptop details 

print(product_name)
print(product_price)
product_rating=product_rating[0:len(product_name)]
print(product_rating)
print(product_details)
```

## *Creating a CSV file of the scrapped data*

```python
# dataframe for CSV file

data={"Name":product_name,"Price":product_price,"Rating":product_rating,"Details":product_details}
```

```python
# CSV file head view

df=pd.DataFrame(data)
df.head()
```

```python
# view CSV file

df=pd.DataFrame(data)
display(df)
```

```python
# writing a CSV file of the Laptop details scrapped

df.to_csv('Laptops_Collection.csv',index=False)
```

# *Creating a SQLite database to store scrapped data*

```python
# Create a connection object,
# Make a new db if not exist already 
# and connect it, if exist then connect.

connection = sq.connect('Laptops_Collection.db') 
```

```python
# Create a cursor object

curs = connection.cursor()
```

```python
# Run create table sql query

curs.execute("create table if not exists Laptops_CollectionInfo" +
             " (Name text, Price integer, Rating integer, Details text)")
```

```python
# Load CSV data into Pandas DataFrame

LAPTOPS = pd.read_csv('Laptops_Collection.csv')
```

```python
# Write the data to a sqlite db table
# Run select sql query

LAPTOPS.to_sql('Laptops_CollectionInfo', connection, if_exists='replace', index=False)
curs.execute('select * from Laptops_CollectionInfo')
```

```python
# Fetch all records as list of tuples

records = curs.fetchall()
```

```python
# Display result 

for row in records:
    # show row
    print(row)
```

```python
# dataframe for database
dataframe = pd.read_sql("""
            SELECT Name, Price, Rating, Details
            FROM Laptops_CollectionInfo
            """, connection)

# return the databse

display(dataframe)
```
